```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Content Based filtering/movies.csv')
```

```python
df.head()
```

```python
df['genres']= df['genres'].str.replace('|',' ')
df['title']= df['title'].str.lower()
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['title']+' '+df['genres'])
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim, top:str=5):
  idx= df[df['title']==title].index[0]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:top+1]
  movie_indices = [i[0] for i in sim_scores]
  return df['title'].iloc[movie_indices], sim_scores
```

```python
movie_name = input('Enter movie name: ')
Top = int(input('Enter top: '))
movie_recommended, similarity_score =get_recommendations(movie_name,cosine_sim,Top)

for i in range(len(movie_recommended)):
  print(f'{i+1}. {movie_recommended.iloc[i]}  Simalirity score: {similarity_score[i][1]}')
```
